```python
import spacy
from spacy import displacy

nlp = spacy.load('en_core_web_lg')
```

```python
import matplotlib.pyplot as plt
```

```python
def visualize_entities(model, text):

    # Create a document
    document = model(text)

    # Show entities in pretty manner
    displacy.render(document, jupyter=True, style='ent')
    #from spacy import displacy
    #displacy.render(doc, style='dep', jupyter=True, options={'distance': 130})
```

```python
def print_entities(pipeline, text):

    # Create a document
    document = pipeline(text)

    # Entity text & label extraction
    for entity in document.ents:
        print(entity.text + '->', entity.label_)

    # Show entities in pretty manner
    #displacy.render(document, jupyter=True, style='ent')
```

```python
with open('/content/PDW_Meeting_Follow_UP_03122024.txt', 'r') as arquivo:
	texto = arquivo.readlines()

for frase in texto:
	if "faturamento" in frase:
		print(frase)
```

```python
import re
```

```python
texto = str(texto)
text = re.sub(r"'", "", texto)
```

```python
print_entities(nlp, text)
```

```python
visualize_entities(nlp, text)
```

```python
from collections import Counter
```

```python
doc = nlp(text)
```

```python
entidades_contagens = Counter([entidade.label_ for entidade in doc.ents])

# Separando os rótulos das entidades e suas contagens
rótulos = list(entidades_contagens.keys())
contagens = list(entidades_contagens.values())

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(rótulos, contagens, color='skyblue')
plt.title('Contagem de Entidades Nomeada - en_core_web_lg')
plt.xlabel('Entidades Nomeadas')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```
